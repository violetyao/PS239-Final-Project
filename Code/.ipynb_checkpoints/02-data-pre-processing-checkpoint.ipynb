```python
import json
import pandas as pd
import jieba
from string import punctuation 
import re
```

load jsons files 

```python
data_path = "data_json/data_train.json"
```

```python
jsons = []
for line in open(data_path, 'r'):
    jsons.append(json.loads(line))
```

```python
new = []
for t in jsons:
    temp = {'fact': t['fact']}
    temp.update(t['meta'])
    new.append(temp)
```

```python
df = pd.DataFrame(new)
```

Melting dataframe

```python
df["term_of_imprisonment"] = df["term_of_imprisonment"].apply(lambda x: eval(x))
```

```python
df['death_penalty'] = df["term_of_imprisonment"].apply(lambda x: x['death_penalty'])
df['imprisonment'] = df["term_of_imprisonment"].apply(lambda x: x['imprisonment'])
df['life_imprisonment'] = df["term_of_imprisonment"].apply(lambda x: x['life_imprisonment'])
```

```python
df["tokenized"] = df["fact"].apply(lambda x: list(jieba.cut(x)))
```

```python
df.to_csv("data_tokenized.csv")
```

remove punctuations and stopwords

```python
def stopwordsList(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding="utf-8").readlines()]
    return stopwords
```

```python
stopwords = stopwordsList("stopwords.txt")
```

```python
df["tokenized"][0]
```

```python
df["cleaned_fact"] = df["tokenized"].apply(lambda fact: [x if x not in stopwords else '' for x in fact]) 
```

```python
add_punc='，。、【】“”：；（）《》‘’{}？！⑦()、%^>℃：.”“^-——=&#@￥'
all_punc=punctuation+add_punc
def sentence_cut(x):
    x = ' '.join(x)
    x=re.sub(r'[A-Za-z0-9]|/d+','',x) #delet numbers and English letters
    testline=x.split(' ')
    te2=[]
    for i in testline:
        te2.append(i)
        if i in all_punc:
            te2.remove(i)
    return te2
```

```python
df["cleaned_fact"] = df["cleaned_fact"].apply(lambda fact: sentence_cut(fact)) 
```
